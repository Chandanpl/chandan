```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=41)
model = AdaBoostClassifier(n_estimators=30, learning_rate=1.0)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print("Predictions:", pred[:5])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

texts = ["free offer now", "win money", "project meeting", "team discussion"]
labels = ["spam", "spam", "ham", "ham"]
cv = CountVectorizer()
X = cv.fit_transform(texts)
nb = MultinomialNB()
nb.fit(X, labels)
test = cv.transform(["free money offer"])
print("Prediction:", nb.predict(test))
```
